Analysing VK Graph with NetworkX
===================================

This notebook contains the analysis of a friends network of my account in VK.
It consists of the following parts
* Network visualization, 
* Saving nodes and edges to JSON format for more stylish visualization libraries
* Performing some basic network analysis

Graph visualization
----------------------------------

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib
import json
from networkx.readwrite import json_graph
%pylab inline
```

```python
g = nx.read_graphml('vk.graphml')
```

We have loaded GraphML, now we can get the nodes positions and draw the graphs. These visualizations are very basic, but we just want to count the node positions for now.

```python
nodes = g.nodes(data=True)

plt.figure(figsize=(19,19), dpi=300,)
positions = {
    'default_fruchterman_reingold': (nx.fruchterman_reingold_layout(g), 221),
    'fruchterman_reingold_100_iterations': (nx.fruchterman_reingold_layout(g, iterations=100), 222),
    'random_layout': (nx.random_layout(g), 223),
    'spring_layout_k_2': (nx.spring_layout(g, iterations=10), 224),
}

for pos in positions:
    plt.subplot(positions[pos][1])
    nx.draw(g, positions[pos][0], with_labels=False)
#plt.show()
plt.savefig("networks.png")
```

Now all we need to save node data to JSON for our JavaScript visualization engine. We will also set color according to the person's sex and deactivated status. We need some additional functions for this:

```python
def get_node_color(node_data):
    if 'deactivated' in node_data['Picture']:
        return "rgb(90,90,90)"
    return "rgb(90,{},90)".format(180 * (int(node_data['Sex'])-1))
```

```python
def export_graph(pos, pos_name):
    edges = [{
                "id": '-'.join(edge), 
                "source": edge[0], 
                "target": edge[1]
            } for edge in g.edges()]

    nodes = [{
                "id": node[0], 
                "label": node[1]['Tooltip'], 
                "color": get_node_color(node[1]),
                "size": 10,
                "x": pos[node[0]][0],
                "y": pos[node[0]][1]
            } for node in g.nodes(data=True)]

    json.dump({"nodes": nodes, "edges": edges}, open('vk_{}.json'.format(pos_name),'w'))
```

Now we are ready to export different graph layouts

```python
for pos in positions:
    export_graph(positions[pos][0], pos)
```

Network analysis
-----------------

First, some basic information about our network:

```python
n_nodes = g.number_of_nodes()
n_edges = g.number_of_edges()

print 'Number of nodes: {}\nNumber of edges: {}'.format(n_nodes, n_edges)
print 'Maximum possible number of edges: {}'.format(n_nodes * (n_nodes - 1)/2)

is_connected = nx.is_connected(g)
print 'Graph is connected: {}'.format(is_connected)
if is_connected:
    print 'Graph diameter: {}\nGraph Center: {}'.format(nx.diameter(g), nx.center(g))
    
print 'Average clustering: {:.3f}'.format(nx.average_clustering(g))
```

```python
subgraphs = nx.connected_component_subgraphs(g)
print 'Number of components in the network: {}'.format(len(subgraphs))
print 'Number of nodes in the main component: {}'.format(subgraphs[0].number_of_nodes())
nx.draw_networkx_edges(subgraphs[0],positions['default_fruchterman_reingold'][0],
                           with_labels=False,
                           edge_color='g',
                           width=6.0)
for subgraph in subgraphs[1:]:
    print 'Number of nodes in the peripheral component: {}'.format(subgraph.number_of_nodes())
    if len(subgraph) > 1:
        nx.draw_networkx_edges(subgraph, positions['default_fruchterman_reingold'][0],
                                     with_labels=False,
                                     edge_color='r',
                                     alpha=0.3,
                                     width=5.0
                                     )

plt.savefig('subgraphs.png')
plt.show()
```

```python
node_data = {node[0]: node[1] for node in g.nodes(data=True)}
centralities = {
                'Betweenness': nx.betweenness_centrality(g),
                'Closeness': nx.closeness_centrality(g),
                'Eigenvector': nx.eigenvector_centrality(g),
                'Degree': nx.degree_centrality(g),
                }
for centrality in centralities:
    for node in sorted(centralities[centrality], key=lambda x: centralities[centrality][x], reverse=True)[:50]:
        #print node, node_data[node]['Tooltip']
        print u'{} centrality for {}: {:.3f}'.format(centrality, node_data[node]['Tooltip'], centralities[centrality][node])
```
